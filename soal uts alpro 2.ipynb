```python
#ngurutin data buat nomor 3
def selectionsort(data) :
  for i in range(len(data)) :
    max_index = i
    for j in range(i+1, len(data)) :
      if data[max_index] > data[j] :
        max_index = j
    data[i], data[max_index] = data[max_index], data[i]
```

```python
# jawaban benar nomor 3 - mean
def mean(data) :
  selectionsort(data) #salahku tadi juga datanya ga diurutin dulu
  median = 0
  if len(data) % 2 == 0 :
    median = (data[int(len(data)/2)-1] + data[int(len(data)/2)]) / 2 #salahku disini, kodeku tdi gak ditambah trus dibagi 2 dulu
  elif len(data) % 2 != 0 :
    median = (data[int(len(data)/2)]) #salahku disini, kodeku tdi malah make +1
  print(data)
  print("Median = ", median)
```

```python
#jawaban benar nomor 3 - modus
def modus(data) :
  jml_nil = []
  for i in range(len(data)) :
    jml = 0
    for j in range(len(data)) :
      if data[i] == data[j] :
        jml += 1
    jml_nil.append(jml) #kayake punyaku salah disini, salah indensasi
  modus = 0
  for k in range(len(jml_nil)) :
    if modus < jml_nil[k]:
      modus = k
  print("Modus : ", data[modus])
```
