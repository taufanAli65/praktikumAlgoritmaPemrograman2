```python
class kendaraan :
  jumlah = 0
  def __init__(kend, merk, warna, tahun) :
    kend.merk = merk
    kend.warna = warna
    kend.tahun = tahun
    kendaraan.jumlah += 1
  def tampil_jumlah(kend) :
    print("Jumlah Kendaraan :", kendaraan.jumlah)
  def tampil_kendaraan(kend) :
    print("Merek :", kend.merk, "| warna :", kend.warna, "| tahun :", kend.tahun)
```

```python
kend1 = kendaraan("Honda", "Hitam", 2004)
kend2 = kendaraan("Kawasaki", "Hitam", 2023)
kend3 = kendaraan("Honda Civic", "Putih", 2020)
kend4 = kendaraan("Yamaha", "Biru", 2023)
```

```python
kend1.tampil_kendaraan()
kend2.tampil_kendaraan()
kend3.tampil_kendaraan()
kend4.tampil_kendaraan()
print("Total :", kendaraan.jumlah)
```
