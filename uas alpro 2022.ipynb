```python
def reverse_number(n):
    r = str(n) #menjadikan n sebagai tipe data string
    return r+r[::-1] #mengembalikan nilai n + n yang sudah direverse

print(reverse_number(1234))
```

```python
def plusminus(n) :
  r = str(n)
  if len(r) > 3 :
    total = int(r[0])
    i = 1
    while i < len(r) :
      if i % 2 == 0 :
        total += int(r[i])
      elif i % 2 != 0 :
        total -= int(r[i])
      i += 1
    return total
  else :
    return "nilai masukkan minimal 3 digit"
print(plusminus(7324512))
```

```python
def pembilangpenyebut(n1, n2) :
  i = 2
  while i <= 9 :
    while n1%i == 0 and n2%i == 0 :
      n1 /= i
      n2 /= i
    i += 1
  return int(n1), int(n2)
print(pembilangpenyebut(9, 24))
```
