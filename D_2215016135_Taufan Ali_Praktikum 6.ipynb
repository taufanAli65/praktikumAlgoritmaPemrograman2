```python
def sequentialsearch(data, key) :
  pos = []
  for i in range (len(data)) :
    if data[i] == key :
      pos.append(i+1)
  if len(pos) >= 1 :
    print("Data", key, "sebanyak", len(pos), "ditemukan di posisi", pos)
  else :
    print("Data tidak ditemukan")
  print("Jumlah iterasi:", i)
```

```python
a = [2,2,1,5,0,1,6,1,3,5]
b = 6
sequentialsearch(a,b)
```

```python
a = "sistem informasi"
b = "i"
sequentialsearch(a,b)
```

```python
a = "taufan ali"
b = "a"
sequentialsearch(a,b)
```

```python
def binsearch(data, key):
    awal = 0
    akhir = len(data) - 1
    posisi = []
    i = 0

    while awal <= akhir:
        tengah = (awal + akhir) // 2
        i += 1
        if key == data[tengah]:
            posisi.append(tengah + 1)
            # Pencarian posisi lainnya ke kiri
            kiri = tengah - 1
            while kiri >= 0 and data[kiri] == key:
                posisi.append(kiri + 1)
                kiri -= 1
            # Pencarian posisi lainnya ke kanan
            kanan = tengah + 1
            while kanan <= akhir and data[kanan] == key:
                posisi.append(kanan + 1)
                kanan += 1
            break
        elif key < data[tengah]:
            akhir = tengah - 1
        else:
            awal = tengah + 1

    if len(posisi) > 0:
        print("Data", key, "ditemukan di posisi:", posisi)
    else:
        print("Data tidak ditemukan")
    print("Jumlah iterasi:", i)
```

```python
a = [1,5,7,8,13,19,20,23,29]
b = 23
binsearch(a,b)
```

```python
a = "sistem informasi"
b = "i"
res = ''.join(sorted(a))
binsearch(res, b)
```

```python
a = "taufan ali"
b = "a"
res = ''.join(sorted(a))
binsearch(res, b)
```
