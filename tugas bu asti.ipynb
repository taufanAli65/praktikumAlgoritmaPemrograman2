```python
def bubblesort(data):
    n = len(data)
    while True:
        tukar = False
        for i in range(n-1):
            if data[i] < data[i+1]:
                data[i], data[i+1] = data[i+1], data[i]
                tukar = True
                print(f"Tukar {data[i]} dengan {data[i+1]}")
                print("urutan menjadi ", data)
        if not tukar:
            break
```

```python
a = [6,1,7,4,2,5]
bubblesort(a)
```
