```python
import math as m

def hitung_bus(peserta, pendamping) :
  total_orang = peserta+pendamping
  total_bus_besar = int(total_orang/46)
  sisa_orang = total_orang % 46
  total_bus_kecil = m.ceil(sisa_orang/16)
  harga_bus_besar = total_bus_besar*800000
  harga_bus_kecil = total_bus_kecil*400000
  total_harga = harga_bus_besar + harga_bus_kecil
  print("Bus Besar : %d"%total_bus_besar)
  print("Bus Kecil : %d"%total_bus_kecil)
  print("Total Biaya Sewa Bus : %d"%total_harga)
```

```python
hitung_bus(250, 7)
```

```python
hitung_bus(300, 10)
```

```python
hitung_bus(350, 12)
```

```python
hitung_bus(0, 0)
```
