```python
import random as rd
def shuffle(array):
    for i in range(len(array)):
        for j in range(len(array[i])):
            i_random = rd.randint(0, 9)
            j_random = rd.randint(0, 9)
            key = array[i][j]
            array[i][j] = array[i_random][j_random]
            array[i_random][j_random] = key
```

```python
array = []

counter = 1
for i in range(0, 10):
    array.append([])
    for j in range(0, 10):
        array[i].append(counter)
        counter += 1
        
shuffle(array)
for row in array:
      print('[', end='')
      for i in range(len(row)):
          if i != len(row) - 1:
              print(row[i], end=',')
          else:
              print(row[i], end='')
      print('],')
```

```python
def radian(derajat) :
  rad = derajat * (3.14/180)
  print(rad)
def derajat(rad) :
  degree = rad * (180/3.14)
  print(degree)
def menu() :
  list_menu = """1. Konversi derajat ke radian
  2. Konversi radian ke derajat"""
  print(list_menu)
  pilih = int(input("Pilih menu [1/2] : "))
  if pilih == 1 :
    a = int(input("Derajat : "))
    radian(a)
  elif pilih == 2 :
    a = int(input("Radian : "))
    derajat(a)
  else :
    menu()
menu()
```

```python
array = [3,4,5,2,1,3,2,2,2,2,3]

def minimum(arr):
    min_val = arr[0]
    for i in range(1, len(arr)):
        if arr[i] < min_val:
            min_val = arr[i]
    print("Min: ", min_val)
    
def median(arr):
    sorted_arr = sorted(arr)
    n = len(sorted_arr)
    if n % 2 == 1:
        median_val = sorted_arr[n//2]
    else:
        median_val = (sorted_arr(n//2)-1) + sorted_arr[n//2]/2
    print("Median: ", median_val)
    
def modus(arr):
    freq = {}
    for num in arr:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    max_freq = max(freq.values())
    modus_val = [num for num, f in freq.items() if f == max_freq]
    print("Modus: ", modus_val)
    return modus_val

minimum(array)
median(array)
modus(array)
```
