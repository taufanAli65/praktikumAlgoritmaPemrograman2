```python
def insertion_sort(data) :
    for i in range(1, len(data)) :
        key = data[i]
        j = i-1
        while j >= 0 and key < data[j] :
            data[j+1] = data[j]
            print("pergeseran pada iterasi ke %d j ke : %d"%(i-1, j), data)
            j -= 1
        data[j+1] = key
        print("iterasi", i-1, data)


#a = [4,3,5,6,2,78,98]
#insertion_sort(a)
```
